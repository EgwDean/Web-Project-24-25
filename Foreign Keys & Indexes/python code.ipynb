```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files for both indexed and non-indexed times
indexed_data = pd.read_csv('times_with_index.csv')  # Replace with the path to your CSV file
non_indexed_data = pd.read_csv('times_without_index.csv')  # Replace with the path to your CSV file

# Plot the graph
plt.plot(indexed_data['iteration'], indexed_data['time_taken'], marker='o', linestyle='-', color='b', label='With Indexes')
plt.plot(non_indexed_data['iteration'], non_indexed_data['time_taken'], marker='o', linestyle='-', color='r', label='Without Indexes')

# Add labels and title
plt.xlabel('Iteration')
plt.ylabel('Time Taken (seconds)')
plt.title('Query Time Comparison: With vs Without Indexes')
plt.legend()

# Save the plot as an image (e.g., PNG format)
plt.savefig('query_time_comparison.png')  # Specify the file name and format

# Display the graph
plt.show()
```
